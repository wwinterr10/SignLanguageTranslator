```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras import models
import numpy as np
```

```python
# One hot encoding for each char in 'hello'
h = [1, 0, 0, 0]
e = [0, 1, 0 ,0]
l = [0, 0, 1, 0]
o = [0, 0, 0, 1]
```

```python
X = np.array([[h, e, l ,l]], dtype="float32")
X
```

```python
#입력값의 줄 칸을 설정하는 Input 객체 선언
#입력값은 4칸임
inputs1 = layers.Input(shape=(4,4))
```

```python
#입력값을 처리하는 LSTM 셀을 구현
whole_sequence_output, final_memory_state, final_result = layers.LSTM(
                                                                    units=1,

                                                                    return_sequences=True,

                                                                    return_state=True
                                                                )(inputs1)
```

```python
#결과는 4칸이 나와야 하기 때문에 4
#활성함수는 softmax
output_dense = layers.Dense(4, activation="softmax") (final_result)
```

```python
#inputs1을 입력받아서 output_dense를 실행하는 모델을 생성
lstm_model = models.Model(inputs1,
                          output_dense
                        )
```

```python
#모델을 실행함 (랜덤한 값 리턴)
lstm_model.predict(X)
```

```python
#학습 방법 설정
lstm_model.compile(
               optimizer="adam", #Adam Optimizer를 이용해서 o를 예측하도록 학습
               loss="categorical_crossentropy", #다중분류
               metrics=["acc"] #학습시 정확도 출력
                )
```

```python
#정답을 저장하고 있는 변수
y = np.array([o], dtype="float32")
y
```

```python
#입력값 X = hell

#정답 y = o

#X 입력시 y를 예측하도록 LSTM, Dense 학습
#eochs=1000 : 1000번 반복해서 학습
lstm_model.fit(X, y, epochs=1000)
```

```python
#학습 후 X 입력
#o = [0,0,0,1] 예측

lstm_model.predict(X)
```
